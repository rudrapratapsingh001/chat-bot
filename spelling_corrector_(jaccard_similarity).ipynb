```python
#correct words
dictionary = ["receive","believe","friend",
               "learning","machine","python",
               "science","data","school",
               "college","project","analysis"]
```

```python
print(dictionary)
```

```python
#convert words into character n-grams (chunks)
def char_ngrams(word,n=2):
  word = word.lower()
  return [word[i:i+n] for i in range(len(word)-n+1)]
```

```python
# jaccard similarity function
def jaccard_similarity(a,b):
  set1=set(a)
  set2=set(b)
  return len(set1 & set2)/len(set1 | set2)
```

```python
jaccard_similarity("apple","apple")
```

```python
#preprocessing to remove numbers and special characters
import re

def preprocessing(word):
  word = word.lower()
  word = re.sub(r'[^a-z]',"",word)
  #removes anything not a-z
  return word
```

```python
preprocessing("rudra@1234")
```

```python
#core spelling correction system
def correct_spelling(word,dictionary,n=2):
  word_ngrams = char_ngrams(word,n)
  best_match = None
  best_score = 0

  for correct_word in dictionary:
    correct_ngrams = char_ngrams(correct_word,n)
    score = jaccard_similarity(word_ngrams,correct_ngrams)

    if score>best_score:
      best_score = score
      best_match = correct_word
  return best_match,best_score
```

```python
#adding a word to a dictionary
dictionary.append("building")
```

#setup and testing

```python
test_words = ["recieve","freind","lerning","machin","scince","builffhhjkjhfd"]
for w in test_words:
  match,score = correct_spelling(w,dictionary)
  print(f"{w}--->{match} (score: {score:.2f})")
```

#user input version


```python
user_word = input("Enter a word :")
match,score = correct_spelling(user_word,dictionary)
print(f"Suggested correction--->{match}(score: {score:2f})")
```
